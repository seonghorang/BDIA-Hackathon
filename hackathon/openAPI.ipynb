```python
### TF-IDFì‚¬ìš©

from sklearn.feature_extraction.text import TfidfVectorizer
from kiwipiepy import Kiwi

text = """ì˜¤ëŠ˜ì€ ë¹„í–‰ê¸°ë¥¼ 5ì‹œê°„ íƒ€ê³  ë² íŠ¸ë‚¨ ë‚˜íŠ¸ëž‘ìœ¼ë¡œ ë– ë‚¬ë‹¤.

ë¹„í–‰ê¸° ì°½ë¬¸ìœ¼ë¡œ ë³´ì´ëŠ” ì•¼ê²½ì´ ì•„ë¦„ë‹¤ì› ë‹¤.

ë¹„í–‰ê¸°ë¥¼ ë‚´ë ¸ë”ë‹ˆ ë”ìœ„ê°€ ë‚˜ë¥¼ ë°˜ê²¼ëŠ”ë° ì •ë§ ì°ì°í–ˆë‹¤.

í˜¸í…”ì´ ë°”ë‹¤ ë°”ë¡œ ì•žì´ë¼ì„œ ë·°ê°€ ê¸°ê°€ ë§‰í˜”ë‹¤.

ë°”ë‹¤ì—ì„œ ìƒì„ ê³¼ ë§¥ì£¼ë¥¼ ë§ˆì…¨ëŠ”ë° ì‹œì›í•˜ê³  ì„¸ìƒì—ì„œ ì œì¼ ë§›ìžˆì—ˆë‹¤.

ìœ ëª…í•œ ê´€ê´‘ì§€ë¥¼ êµ¬ê²½í•˜ê³  í…Œë§ˆíŒŒí¬ë¡œ ì´ë£¨ì–´ì§„ ì„¬ì„ ê°”ëŠ”ë°, ì•¼ìžìˆ˜ê°€ ì´ë»¤ë‹¤

ì›Œí„°íŒŒí¬ì—ì„œ ì‹ ëª…ë‚˜ê²Œ ë†€ì•˜ë‹¤."""

# Kiwi ê°ì²´ ìƒì„±
kiwi = Kiwi()

# ë¬¸ìž¥ì—ì„œ ëª…ì‚¬ ì¶”ì¶œ
nouns_list = []
for sentence in kiwi.analyze(text):
    nouns = [token.form for token in sentence[0] if token.tag.startswith('NN')]
    if nouns:
        nouns_list.extend(nouns)

# TfidfVectorizer ê°ì²´ ìƒì„±
vectorizer = TfidfVectorizer()

# ëª…ì‚¬ë“¤ë¡œ êµ¬ì„±ëœ ë¬¸ì„œ-ë‹¨ì–´ í–‰ë ¬ ìƒì„±
X = vectorizer.fit_transform(nouns_list)

# ê° ë‹¨ì–´ì˜ TF-IDF ì ìˆ˜ ê³„ì‚°
feature_names = vectorizer.get_feature_names_out()
tfidf_scores = {word: tfidf for word, tfidf in zip(feature_names, X.toarray().sum(axis=0))}

# TF-IDF ì ìˆ˜ê°€ ë†’ì€ ìˆœìœ¼ë¡œ ì •ë ¬ (ìƒìœ„ 20ê°œ)
keywords = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:20]

print(keywords)
```
