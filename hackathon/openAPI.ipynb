```python
from keybert import KeyBERT
from kiwipiepy import Kiwi
from transformers import BertModel

model = BertModel.from_pretrained('skt/kobert-base-v1')
kw_model = KeyBERT(model)

text = """오늘은 비행기를 5시간 타고 베트남 나트랑으로 떠났다.

비행기 창문으로 보이는 야경이 아름다웠다.

도착 후 비행기를 내리니 더위와 습기가 나를 반겼는데 정말 찝찝했다.

호텔이 해변 바로 앞이라서 뷰가 기가 막혔다.

해변에서 닭꼬치와 맥주를 마셨는데 시원하고 세상에서 제일 맛있었다.

유명한 관광지를 구경하고 빈원랜드라는 테마파크로 이루어진 섬을 갔는데,

놀이기구와 워터파크에서 신명나게 놀았다.

다음에 베트남 또 가기로 했다."""

kiwi = Kiwi()
nouns_list = []
for sentence in kiwi.analyze(text):
    nouns = [token.form for token in sentence[0] if token.tag.startswith('NN')]
    if nouns:
        nouns_list.extend(nouns)
result_text = ' '.join(nouns_list)

keywords = kw_model.extract_keywords(result_text, keyphrase_ngram_range=(1, 1), stop_words=None, top_n=20)
print(keywords)
```

```python
### TF-IDF사용

from sklearn.feature_extraction.text import TfidfVectorizer
from kiwipiepy import Kiwi

text = """오늘은 비행기를 5시간 타고 베트남 나트랑으로 떠났다.

비행기 창문으로 보이는 야경이 아름다웠다.

비행기를 내렸더니 더위가 나를 반겼는데 정말 찝찝했다.

호텔이 바다 바로 앞이라서 뷰가 기가 막혔다.

바다에서 생선과 맥주를 마셨는데 시원하고 세상에서 제일 맛있었다.

유명한 관광지를 구경하고 테마파크로 이루어진 섬을 갔는데, 야자수가 이뻤다

워터파크에서 신명나게 놀았다."""

# Kiwi 객체 생성
kiwi = Kiwi()

# 문장에서 명사 추출
nouns_list = []
for sentence in kiwi.analyze(text):
    nouns = [token.form for token in sentence[0] if token.tag.startswith('NN')]
    if nouns:
        nouns_list.extend(nouns)

# TfidfVectorizer 객체 생성
vectorizer = TfidfVectorizer()

# 명사들로 구성된 문서-단어 행렬 생성
X = vectorizer.fit_transform(nouns_list)

# 각 단어의 TF-IDF 점수 계산
feature_names = vectorizer.get_feature_names_out()
tfidf_scores = {word: tfidf for word, tfidf in zip(feature_names, X.toarray().sum(axis=0))}

# TF-IDF 점수가 높은 순으로 정렬 (상위 20개)
keywords = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:20]

print(keywords)
```
